**a)** $W$ which is uniformly distributed over the interval $[a, b]$, for some value $a,b \in \mathbb{R}$.

```r
# Your Code Here
options(repr.plot.width = 15, repr.plot.height = 15)

f <- function(x) {
    return (ifelse((x >= a) && (x <= b), 1 / (b - a), 0))
}
f <- Vectorize(f)

plot_graph <- function(x, f, title) {
    ggplot() + geom_line(aes(x, f(x))) + 
               geom_point(aes(median, f(median), col='median'), size=5) +
               geom_point(aes(mode, f(mode), col='mode'), size=9, alpha=0.6) +
               geom_point(aes(mean, f(mean), col='mean'), size=15, alpha=0.3) + 
               labs(col='measures of\ncentral tendency') +
               #scale_shape(solid = FALSE) +
               theme(text = element_text(size=20)) + ggtitle(title)
}

# parameters
a <- 0
b <- 1

median <- (a+b)/2
mode <- (a+b)/2 # can be any point in between [a,b], just one is shown in the figure
mean <- (a+b)/2

print(c(median, mode, mean))

library(ggplot2)

x <- seq(a-0.5,b+0.5,0.01)
plot_graph(x, f, 'Uniform Distribution')
```

         \frac{1}{b-a}, & \text{for } a\leq x\leq b\\
        0, & \text{otherwise}
        \end{array}\right\}$

(ii) 

\begin{array}
\text{Median: } & F(x) = \int\limits_{a}^{x}\frac{1}{b-a}dx = \left[\frac{x}{b-a}\right]_a^x=\frac{x-a}{b-a}. \text{ Now } F(x_{m})=\frac{x_{m}-a}{b-a}=0.5 \implies x_m=\frac{a+b}{2} \\
\text{Mode: } & M = \underset{a\leq x\leq b}{argmax} \frac{1}{b-a} = \{x: a \leq x \leq b\}, i.e., \text{ any point in between $a$ and $b$, since $f(x)$ is a constant function} \\
\text{Mean: } & \mu=\int\limits_{a}{b}\frac{x}{b-a}dx=\frac{1}{b-a}\left[\frac{x^2}{2}\right]_a^b=\frac{a+b}{2} \\
\end{array}
