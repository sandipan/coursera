# K-Means and K-Medoid
We learned both K-Means and K-Medoid as basic clustering methods. Let's play with them and observe the differences.

## Prepare the packages


*   K-Means is included in sklearn package, we can import it directly.
*   K-Medoid is included in sklearn extra package, we should install it first, then import it. 
*   We will need numpy and pandas to create some dummy dataset
*   We will need seaborn for visualize the result.

```python
!pip install scikit-learn-extra
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
import seaborn as sns
```

# Round 1: Let's create some simple dataset.
For visualization purpose, we create datapoints with 2 dimensions.

### Create 3 clusters of data points.
Each cluster has 1000 data points, each data point has 2 dimensions. Data points has normal distribution with specified mean and std.

```python
c1 = np.random.normal(5, 3, (100, 2))
c2 = np.random.normal(15, 5, (100, 2))
c3 = np.random.normal(-5, 2, (100, 2))
d = np.concatenate((c1, c2, c3), axis = 0)
d.shape
```

### Let's visualize the dataset using Seaborn

```python
sns.scatterplot(data = d, x = d[:,0], y = d[:,1])
```

### Let's train a model using KMeans

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(d)
```

```python
sns.scatterplot(data = d, x = d[:,0], y = d[:,1],hue = kmeans.predict(d))
```

### Lets train a model using KMedoid

```python
kmedoids = KMedoids(n_clusters=3, random_state=0).fit(d)
```

### Visualize the result

```python
sns.scatterplot(data = d, x = d[:,0], y = d[:,1],hue = kmedoids.predict(d))
```

## Conclusion: KMean and KMedoid did equaly well for this dataset.

#Round 2: Let us add one outlier to the dataset


### Create the same random clusters

```python
c1 = np.random.normal(5, 3, (100, 2))
c2 = np.random.normal(15, 5, (100, 2))
c3 = np.random.normal(-5, 2, (100, 2))
```

The outlier is far away from all other data points


```python
outlier = np.array([[100, 100]])
```

### Train a model using KMedoids

```python
kmedoids = KMedoids(n_clusters=3, random_state=0).fit(d)
```

###visualize the result

```python
sns.scatterplot(data = d, x = d[:,0], y = d[:,1],hue = kmedoids.predict(d))
```

### Conclusion: KMedoid did well even with an outlier. KMeans failed

#Round 3: Let's create a more realistic dataset using pandas

## create a dataset

```python
df = pd.DataFrame({'x': np.random.normal(5, 3, (100)), 'y': np.random.normal(-2, 2, (100))})
df = df.append(pd.DataFrame({'x': np.random.normal(15, 2, (100)), 'y': np.random.normal(22, 2, (100))}))
df = df.append(pd.DataFrame({'x': np.random.normal(-5, 3, (100)), 'y': np.random.normal(8, 2, (100))}))
```

### Visualize the result

### Train a model using KMdoids

```python
kmedoids = KMedoids(n_clusters=3, random_state=0).fit(df)
```

### Visualize the result

```python
sns.scatterplot(data = df, x = 'x', y = 'y', hue = kmedoids.predict(df))
```
